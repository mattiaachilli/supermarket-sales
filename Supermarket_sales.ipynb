# Cleaning and transform data

Dropping useless column: we can remove Invoice ID because is not useful for our analysis (is just a identifier of the sale)

```python
df.drop(columns=["Invoice ID", "gross margin percentage"], inplace=True)
```

```python
df.columns
```

We can concat date and time in a unique column

```python
# Concat date and time
df["Date"] = df["Date"].astype(str) + " " + df["Time"].astype(str)
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.drop(columns=["Time"], inplace=True)
```

# EDA (Exploratory data analysis)

We can get the percentage of each rating on the total (distribution of ratings)

```python
pd.set_option("display.max_rows", None)
df["Rating"].value_counts(normalize=True) * 100
```

Better with a piechart, even if there are too many values

```python
df["Rating"].value_counts().plot.pie();
```

Or histogram

```python
plt.hist(df["Rating"], bins=df["Rating"].value_counts().count())
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Distribution of ratings")
plt.show()
```

Now we can see value counts (piechart and also hist) for each categorical column.

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

# All columns seem to have same distribution

fig = plt.figure(figsize=(12, 6))

for column in categorical_columns:
  plt.subplot(2, 3, categorical_columns.index(column) + 1)
  df[column].value_counts().plot(kind='pie', autopct='%1.1f%%')
  plt.title(column)
```

A similar chart for see the distribution

```python
fig = plt.figure(figsize=(12, 6))

for column in categorical_columns:
  plt.subplot(2, 3, categorical_columns.index(column) + 1)
  plt.hist(df[column])
```

Let see also the distribution of numerical data.

Obviously here for numerical data there are too many unique values but in certain dataset this is also interesting

```python
numerical_columns = df.select_dtypes(include=['float64', "int64"]).columns.tolist()

fig = plt.figure(figsize=(30, 15))

for column in numerical_columns:
  plt.subplot(2, 4, numerical_columns.index(column) + 1)
  plt.hist(df[column], bins=df[column].value_counts().count())
  plt.title(column)
```

Now we can try to see correlation between columns, in this task we should find the column that are most related to rating column, this is a regression task for ML

We need to convert categorical data into numerical for correlation matrix

This graph let us to know which column is much related to Rating column, we can see some correlations like Total to Tax 5%

```python
corr = df[numerical_columns].corr()
sns.heatmap(corr, cmap="Blues", annot=True)
```

Obviously when total increase tax increase, is linear

Encoding all categorical columns in order to get a value for each label in the column, ex: A => 0, B => 1, and so on... for this there are differente techniques, it depends also by the alghoritm of ML we want to use

```python
for column in categorical_columns:
  df[column] = df[column].astype('category')
  df[column] = df[column].cat.codes
```

In this way we can see correlation with all columns

```python
plt.figure(figsize=(30, 15))
sns.heatmap(df.corr(), cmap="Blues", annot=True, linewidths=1, linecolor="black")
```

Then we could predict the rating based feature, using different alghoritms of ML like linear regression (not in this case), decision tree, random forest, SVR, DNN (deep neural networks) and maybe measuring RSME.
